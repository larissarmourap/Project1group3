```python
# adding a new column "Woman_ID" with unique identifiers starting from 1
maternal_health_df['Woman_ID'] = range(1,len(maternal_health_df)+1)
                                   
#Reordering the DataFrame columns
columns = ['Woman_ID']+[col for col in maternal_health_df.columns if col !='Woman_ID']
maternal_health_df = maternal_health_df[columns]

# Print
maternal_health_df.head()
```
